```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_excel("./provinces_updated.xls")
df = df.sort_values('Population')
df
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

NUMBER = [x for x in range(1, 64)]
plt.bar(NUMBER, df['Population'], width = 0.7)

# plt.xlim(10, 100)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.suptitle('Population in Vietnamese Cities & Provinces', fontsize=24, color='#444444', x=0.125, ha='left')
plt.title('Sorted ascending', fontsize=18, color='#777777', x=0, ha='left')
plt.xlabel('Cities', fontsize=18, color='#333333')
plt.ylabel('Population (million)', fontsize=18, color='#333333')

plt.show()
```

```python
for i in range(0, len(NUMBER)):
    print(f"{NUMBER[i]}."+ df.iloc[i]['Name'], end = '  ')
```

```python
reg_df = pd.read_excel("./region_areas.xls")
reg_df
```

```python
colors = ['#306dbf', '#ff3300', '#00cc66', '#cc0000', '#cc33ff', '#996633', '#ff99cc' , '#a6a6a6']
plt.pie(reg_df['Area Proportion (%)'], autopct='%1.1f%%', startangle=90, counterclock=False,
colors=colors, wedgeprops={'edgecolor': 'w', 'linewidth': 1}, labels = reg_df['Region'])
plt.title('Proportion of Area in Regions', fontsize=16, color='#333333', x=0.125, ha='left')
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

n, bins, edges = plt.hist(df['PopulationDensity'], bins=22, ec='white')
plt.xticks(bins)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.suptitle('Frequency of Population Density in Vietnam', fontsize=24, color='#444444', x=0.125, ha='left')
plt.xlabel('Population Density (people/km2)', fontsize=18, color='#333333')
plt.ylabel('Frequency', fontsize=18, color='#333333')
plt.show()
```
