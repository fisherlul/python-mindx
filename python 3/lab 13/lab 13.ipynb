```python
import matplotlib.pyplot as plt 
import pandas as pd
import datetime as dt
```

```python
data = pd.read_csv('covid19.csv')
data
```

```python
df = pd.read_csv('population.csv')
df
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(data['date'], data['confirmed'], width = 0.8)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.xticks([])
plt.yticks([])

confirmed_cases = data['confirmed'].sum()

plt.suptitle('Total Confirmed Cases', fontsize=24, color='#a6a6a6', x=0.125, ha='left')
plt.title("{:,}".format(confirmed_cases), fontsize=16, color='#a6a6a6', x=0, ha='left')
# plt.xlabel('Date', fontsize=18, color='#333333')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(data['date'], data['deaths'], width = 0.9)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.xticks([])
plt.yticks([])
plt.ylim(0, max(data['deaths']))

dead_cases = data['deaths'].sum()

plt.suptitle('Total Deaths', fontsize=24, color='#a6a6a6', x=0.125, ha='left')
plt.title("{:,}".format(dead_cases), fontsize=16, color='#a6a6a6', x=0, ha='left')

plt.show()
```
